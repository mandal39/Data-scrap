```python
import requests
import urllib.request
from bs4 import BeautifulSoup
```

```python
baseurl = 'https://www.zangia.mn/job/list/pg.'
app_list = []
for i in range(1,161): 
  url = baseurl + str(i)
  #print(url)
  #print(url)
  response = requests.get(url)
  if response.status_code != 200:
   print(response.status_code)
   print('error ',url)
   continue
  print(i)
  soup = BeautifulSoup(response.text,"html.parser")
  li_list = soup.find_all("div", {"class": "ad"})
  for li in li_list:
    a = li.find('a')
    work_url = 'https://www.zangia.mn/'+a['href']
    #print(work_url)
    app_list.append(work_url)
```

```python
print(len(app_list))
app_set = set(app_list)
print(len(app_set))
```

```python
class Work:
  def __init__(self, company, title, location, salary, jtype, level, branch, role, requirements, date, url):
    self.company = company
    self.title = title
    self.location = location
    self.salary = salary
    self.jtype = jtype
    self.level = level
    self.branch = branch
    self.role = role
    self.requirements = requirements
    self.date = date
    self.url = url
```

```python
def findFeature(li_list, header):
  ret = 'NA'
  for li in li_list:
    text = li.text.strip()
    if text.startswith(header):
      return text[len(header) + 1:]
  return ret
```

```python
it = 0
data_list = []
for url in app_set: 
  #print(url)
  it += 1
  if it > 7999:
    break
  print(url)
  response = requests.get(url)
  if response.status_code != 200:
   print(response.status_code)
   print('error ',url)
   continue
  soup = BeautifulSoup(response.text,"html.parser")
  title = soup.find("h3").text.strip()
  company = soup.find('td')
  print(type(company))
  if(company != None):
    if(bool(company.find('p'))==True):
      c = company.p.decompose()
      company = company.text
  div_class = soup.find_all("div")
  salary = findFeature(div_class,'Цали').strip()
  jtype = findFeature(div_class,'Төрө').strip()
  location = findFeature(div_class,'Байрши').strip()
  branch = findFeature(div_class, 'Салба').strip()
  level = findFeature(div_class, 'Түвши').strip()
  role = findFeature(div_class, 'Гүйцэтгэх үндсэн үүрэ').strip()
  requirements = findFeature(div_class, 'Ажлын байранд тавигдах шаардлаг').strip()
  date = findFeature(div_class, 'Зар нийтэлсэн огно').strip()
  rloc = ""
  rlocation = ""
  for loc in location:
    rlocation = rlocation + loc
    if(loc.isupper()==True):
      rloc = loc
    else:
      rloc = rloc + loc
      if(rloc=="Салбар"):
        break
  location = rlocation[0: len(rlocation)-6]
  data = (title, company, branch, level, location, salary, jtype, role, requirements, date)
  data_list.append(data)
  
```
